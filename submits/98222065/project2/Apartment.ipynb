```python
min_max_scaler = MinMaxScaler()
```

```python
x_train = min_max_scaler.fit_transform(x_train)
x_test = min_max_scaler.transform(x_test)
```

```python
class LiniearRegression_MSE() :   
    def __init__(self, alpha=0.05 , iterations=1000):
        self.alpha = alpha
        self.iter = iterations 
    
    def fit(self, x, y): 
        self.m, self.n = x.shape 
        self.W = np.random.normal(loc=0.0, scale=0.001, size=self.n)
        self.b = 0

        for i in range(self.iter) :
            y_pred = self.predict(x)
            error = y - y_pred
            dW = -(2 * (x.T).dot(error)) / self.m
            db = -2 * np.sum(error) / self.m
            self.W = self.W - self.alpha * dW
            self.b = self.b - self.alpha * db 
        return self

    def predict(self, x) :
        return (x.dot( self.W ) + self.b)
```

```python
liniearregression_mse = LiniearRegression_MSE(alpha=0.1)
```

```python
liniearregression_mse.fit(x_train, y_train)
```

```python
y_pred_mse = liniearregression_mse.predict(x_test)
```

```python
from sklearn import metrics
metrics.mean_squared_error(y_test, y_pred_mse)
```

```python
from sklearn.linear_model import LinearRegression, Lasso, Ridge
```

```python
linear_regression = LinearRegression()
```

```python
linear_regression.fit(x_train, y_train)
```

```python
y_pred_lr = linear_regression.predict(x_test)
```

```python
metrics.mean_squared_error(y_test, y_pred_lr)
```

```python
metrics.r2_score(y_test, y_pred_rdg)
```

```python
lasso = Lasso(alpha=0.1)
lasso.fit(x_train, y_train)
y_pred_las = lasso.predict(x_test)
```

```python
metrics.r2_score(y_test, y_pred_las)
```
